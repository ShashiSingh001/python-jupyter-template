### Exercise 6: Histogram Customization

```python
# TASK: Create a histogram for the following data: data = [1, 2, 2, 3, 3, 3, 4, 4, 4, 4, 5].
# Customize the histogram with a title, labels for the x-axis, and a specific color for the bars.
import matplotlib.pyplot as plt
import numpy as np

# Create a 2x2 subplot layout
fig, axs = plt.subplots(2, 2, figsize=(10, 10))

# Line chart
t = np.linspace(0, 10, 100)
y = np.sin(t)
axs[0, 0].plot(t, y, color='blue')
axs[0, 0].set_title('Line Chart')
axs[0, 0].grid(True)

# Bar chart
categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 9]
axs[0, 1].bar(categories, values, color=['red', 'blue', 'green', 'purple'])
axs[0, 1].set_title('Bar Chart')

# Scatter plot
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]
axs[1, 0].scatter(x, y, color='blue')
for i, j in zip(x, y):
    axs[1, 0].text(i, j, f'({i}, {j})', fontsize=12, ha='right')
axs[1, 0].set_title('Scatter Plot')
axs[1, 0].grid(True)

# Pie chart
labels = ['Python', 'Java', 'C++', 'JavaScript']
sizes = [40, 25, 20, 15]
axs[1, 1].pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140, colors=['blue', 'red', 'green', 'orange'])
axs[1, 1].set_title('Pie Chart')

# Histogram
fig_hist, ax_hist = plt.subplots(figsize=(8, 5))
data = [1, 2, 2, 3, 3, 3, 4, 4, 4, 4, 5]
ax_hist.hist(data, bins=5, color='cyan', edgecolor='black')
ax_hist.set_title('Histogram')
ax_hist.set_xlabel('Value')
ax_hist.set_ylabel('Frequency')
ax_hist.grid(axis='y', linestyle='--', alpha=0.7)

# Adjust layout
plt.tight_layout()

# Show the plots
plt.show()
```
